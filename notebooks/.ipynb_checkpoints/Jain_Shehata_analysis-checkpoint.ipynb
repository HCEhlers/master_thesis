# Loading FASTA-sequences for each dataset

```python
fasta_VL_jain = {
    fasta.id: str(fasta.seq)
    for fasta in SeqIO.parse(
        "data/processed/data_extracted_jain_VL_aligned.fasta", "fasta"
    )
}
fasta_VH_jain = {
    fasta.id: str(fasta.seq)
    for fasta in SeqIO.parse(
        "data/processed/data_extracted_jain_VH_aligned.fasta", "fasta"
    )
}
fasta_VL_shehata = {
    fasta.id: str(fasta.seq)
    for fasta in SeqIO.parse(
        "data/processed/data_extracted_shehata_VL_aligned.fasta", "fasta"
    )
}
fasta_VH_shehata = {
    fasta.id: str(fasta.seq)
    for fasta in SeqIO.parse(
        "data/processed/data_extracted_shehata_VH_aligned.fasta", "fasta"
    )
}
```

## Calculating sequence ID between all pairs in the datasets

```python
jain_VL = getSeqIDmat(list(fasta_VL_jain.values()))
jain_VH = getSeqIDmat(list(fasta_VH_jain.values()))
jain_indices = np.triu_indices_from(jain_VL, k=1)
jain_VL = jain_VL[jain_indices]
jain_VH = jain_VH[jain_indices]
shehata_VL = getSeqIDmat(list(fasta_VL_shehata.values()))
shehata_VH = getSeqIDmat(list(fasta_VH_shehata.values()))
shehata_indices = np.triu_indices_from(shehata_VL, k=1)
shehata_VL = shehata_VL[shehata_indices]
shehata_VH = shehata_VH[shehata_indices]
combined_VL = getSeqIDmat(list(fasta_VL_jain.values()) + list(fasta_VL_shehata.values()))
combined_VH = getSeqIDmat(list(fasta_VH_jain.values()) + list(fasta_VH_shehata.values()))
combined_indices = np.triu_indices_from(combined_VL, k=1)
combined_VL = combined_VL[combined_indices]
combined_VH = combined_VH[combined_indices]
```

```python
jain_id_data = {"VL": jain_VL, "VH": jain_VH}
jain_id_df = pd.DataFrame.from_dict(jain_id_data)
jain_id_df["Dataset"] = "Jain"
shehata_id_data = {"VL": shehata_VL, "VH": shehata_VH}
shehata_id_df = pd.DataFrame.from_dict(shehata_id_data)
shehata_id_df["Dataset"] = "Shehata"
combined_id_data = {"VL": combined_VL, "VH": combined_VH}
combined_id_df = pd.DataFrame.from_dict(combined_id_data)
combined_id_df["Dataset"] = "Combined"
id_df = pd.concat([jain_id_df, shehata_id_df, combined_id_df])
```

## Obtaining quantiles for sequence identity for both VL and VH chains

```python
stat_id_VL = (
    id_df.groupby(["Dataset"])
    .describe()["VL"][["mean", "min", "25%", "50%", "75%", "max"]]
    .round(4)
)
#combined_id_VL = (
#    id_df["VL"]
#    .describe()[["mean", "min", "25%", "50%", "75%", "max"]]
#    .T
#    .round(4)
#)
#id_VL = pd.concat([stat_id_VL, combined_id_VL])
#id_VL.index = ["Jain (VL)", "Shehata (VL)", "Combined (VL)"]
stat_id_VL
```

```python
stat_id_VH = (
    id_df.groupby(["Dataset"])
    .describe()["VH"][["mean", "min", "25%", "50%", "75%", "max"]]
    .round(4)
)
#combined_id_VH = (
#    id_df["VH"]
#    .describe()[["mean", "min", "25%", "50%", "75%", "max"]]
#    .to_frame()
#    .T.round(4)
#)
#id_VH = pd.concat([stat_id_VH, combined_id_VH])
#id_VH.index = ["Jain (VH)", "Shehata (VH)", "Combined (VH)"]
stat_id_VH
```

```python
stat_id = (
    id_df.groupby(["Dataset"])
    .describe()
    .loc[:, (slice(None), ["mean", "min", "25%", "50%", "75%", "max"])]
)
stat_id.loc["Combined"] = pd.concat(
    [
        id_df.VL.describe()[["mean", "min", "25%", "50%", "75%", "max"]],
        id_df.VH.describe()[["mean", "min", "25%", "50%", "75%", "max"]],
    ]
).to_numpy()
stat_id[::-1]
```

```python
print(
    stat_id[::-1].round(3).T.to_latex(
        label="tab:seqIDStat", caption="Sequence identity statistics of datasets"
    )
)
```

```python
sns.histplot(
    id_df["VL"][id_df.Dataset == "Jain"], label="Jain", color="red", kde=True, alpha=1.0
)
sns.histplot(
    id_df["VL"][id_df.Dataset == "Shehata"],
    label="Shehata",
    color="grey",
    kde=True,
    alpha=0.8,
)
sns.histplot(id_df["VL"], label="Combined", color="pink", kde=True, alpha=0.4)
plt.legend()
plt.xlabel("Sequence identity")
plt.ylabel("Count")
plt.savefig("reports/figures/VL_seqID_distribution.png")
plt.show()
```

```python
sns.histplot(
    id_df["VH"][id_df.Dataset == "Jain"], label="Jain", color="red", kde=True, alpha=1.0
)
sns.histplot(
    id_df["VH"][id_df.Dataset == "Shehata"],
    label="Shehata",
    color="grey",
    kde=True,
    alpha=0.8,
)
sns.histplot(id_df["VH"], label="Combined", color="pink", kde=True, alpha=0.4)
plt.legend()
plt.xlabel("Sequence identity")
plt.ylabel("Count")
plt.savefig("reports/figures/VH_seqID_distribution.png")
plt.show()
```

## Creating heatmaps

```python
# Extracting sequence ID matrix
names = list(fasta_VL_jain.keys()) + list(fasta_VL_shehata.keys())
dataset = ["Shehata" if name.startswith("ADI_") else "Jain" for name in names]
seq_VL = list(fasta_VL_jain.values()) + list(fasta_VL_shehata.values())
seq_VH = list(fasta_VH_jain.values()) + list(fasta_VH_shehata.values())
seq_VL_id = getSeqIDmat(seq_VL)
seq_VH_id = getSeqIDmat(seq_VH)
```

```python
a = dict(zip(np.unique(dataset), "rg"))
colors = lambda x: a[x]
row_colors = list(map(colors, dataset))
```

```python
g = sns.clustermap(seq_VL_id, row_colors=row_colors, yticklabels=False,xticklabels=False)
new_labels = []
for l in g.ax_heatmap.axes.get_xticklabels():
    tick_text = l.get_text()
    #l.set_text(names[int(tick_text)])
    DSname = dataset[int(tick_text)]
    l.set_color(a[DSname])
    new_labels.append(l)
g.ax_heatmap.axes.set_xticklabels(new_labels)
new_labels = []
for l in g.ax_heatmap.axes.get_yticklabels():
    tick_text = l.get_text()
    #l.set_text(names[int(tick_text)])
    DSname = dataset[int(tick_text)]
    l.set_color(a[DSname])
    new_labels.append(l)
g.ax_heatmap.axes.set_yticklabels(new_labels)

plt.savefig("reports/figures/VL_heatmap.png")
plt.show()
```

```python
g = sns.clustermap(seq_VH_id, row_colors=row_colors, yticklabels=False,xticklabels=False)
new_labels = []
for l in g.ax_heatmap.axes.get_xticklabels():
    tick_text = l.get_text()
    l.set_text(names[int(tick_text)])
    DSname = dataset[int(tick_text)]
    l.set_color(a[DSname])
    new_labels.append(l)
g.ax_heatmap.axes.set_xticklabels(new_labels)
new_labels = []
for l in g.ax_heatmap.axes.get_yticklabels():
    tick_text = l.get_text()
    l.set_text(names[int(tick_text)])
    DSname = dataset[int(tick_text)]
    l.set_color(a[DSname])
    new_labels.append(l)
g.ax_heatmap.axes.set_yticklabels(new_labels)
plt.savefig("reports/figures/VH_heatmap.png")
plt.show()
```
