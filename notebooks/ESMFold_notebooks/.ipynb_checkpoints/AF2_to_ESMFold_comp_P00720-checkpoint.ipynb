```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
os.chdir('../..')
cwd = os.getcwd()
DATA_PATH_ROOT = cwd
# Load data
DATA_PATH_MUT_ESM = DATA_PATH_ROOT + '/data/results_rmsd/results_rmsd_esmfold/P00720/P00720_rec1_AF2/'
DATA_PATH_MUT_ESM_rec3 = DATA_PATH_ROOT + '/data/results_rmsd/results_rmsd_esmfold/P00720/P00720_rec3_Af2/'

arodz_pdbs = ['1dya', '1dyb', '1dyc', '1dyd', '1dye', '1dyf', '1dyg', '1l00',
       '1l02', '1l03', '1l05', '1l06', '1l07', '1l08', '1l09', '1l10',
       '1l11', '1l12', '1l13', '1l14', '1l15', '1l16', '1l19', '1l20',
       '1l21', '1l22', '1l23', '1l24', '1l25', '1l26', '1l27', '1l28',
       '1l29', '1l30', '1l31', '1l32', '1l33', '1l34', '1l37', '1l38',
       '1l44', '1l45', '1l46', '1l47', '1l48', '1l52', '1l53', '1l56',
       '1l57', '1l58', '1l60', '1l69', '1l98', '1l99']
# ESMFold muts to 5 AF2
af2_wt_rmsd_list = np.zeros((5,54))
dir_list = sorted(os.listdir(DATA_PATH_MUT_ESM))
for j in range(len(dir_list)):
    df_tmp = pd.read_csv(DATA_PATH_MUT_ESM + dir_list[j])['name']
    df_tmp2 = pd.read_csv(DATA_PATH_MUT_ESM + dir_list[j])
    df_tmp2['model_n'] = list(df_tmp.str.extract(r'(model_)(\d)').iloc[:,1])
    df_tmp2 = df_tmp2.sort_values('model_n')
    ls = np.asarray(list(df_tmp2['rmsd']))
    for idx in range(af2_wt_rmsd_list.shape[0]):
        af2_wt_rmsd_list[idx,j] = ls[idx]

af2_wt_rmsd_list = np.asarray(af2_wt_rmsd_list)
```

```python
af2_wt_rmsd_list
np.mean(af2_wt_rmsd_list, axis=1)
```

```python
# Create lists for the plot
uniprot_ids = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']
x_pos = np.arange(len(uniprot_ids))
means = np.mean(af2_wt_rmsd_list, axis=1).tolist()
errors = np.std(af2_wt_rmsd_list, axis=1).tolist()
errors
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, means, yerr=errors, align='center', alpha=1, ecolor='black', capsize=10)
ax.set_ylabel('RMSD [Å]')
ax.set_xticks(x_pos)
ax.set_xticklabels(uniprot_ids)
ax.set_title("P00720: Mean RMSD between ESMFold Model and AF2's 5 Models")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig(DATA_PATH_ROOT + '/notebooks/figs/bar_plot_P00720_ESM_AF2.png')
plt.show()
```
