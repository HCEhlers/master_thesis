```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
# Load data
DATA_PATH_MUT_ESM_rec1 = '/Users/holger/Desktop/master_thesis/data/results_rmsd/results_rmsd_esmfold/P61626/P61626_rec1_Af2/'
DATA_PATH_MUT_ESM_rec3 = '/Users/holger/Desktop/master_thesis/data/results_rmsd/results_rmsd_esmfold/P61626/P61626_rec3_Af2/'
arodz_pdbs = ['1gb5', '1gb6', '1gb7', '1gb8', '1gb9', '1gbo', '1gbw', '1gbx', '1gby', '1gbz', '1gfh', '1gfj', '1gfk', '1gfr', '1gft', '1gfu','1gfv', '1inu', '1lhh', '1lhi', '1lhj', '1lhk', '1lhl', '1oub','1ouc', '1oud', '1oue', '1ouh', '1oui', '1ouj', '1wqm', '1wqn','1wqo', '1wqp', '1wqq', '1wqr', '1yam', '1yan', '1yao', '1yap','1yaq', '2hea', '2heb', '2hec', '2hed', '2hee', '2hef']

# ESMFold muts to 5 AF2
af2_wt_rmsd_list_rec1 = np.zeros((5,47))
dir_list = sorted(os.listdir(DATA_PATH_MUT_ESM_rec1))
for j in range(len(dir_list)):
    df_tmp = pd.read_csv(DATA_PATH_MUT_ESM_rec1 + dir_list[j])['name']
    df_tmp2 = pd.read_csv(DATA_PATH_MUT_ESM_rec1 + dir_list[j])
    df_tmp2['model_n'] = list(df_tmp.str.extract(r'(model_)(\d)').iloc[:,1])
    df_tmp2 = df_tmp2.sort_values('model_n')
    ls = np.asarray(list(df_tmp2['rmsd']))
    for idx in range(af2_wt_rmsd_list_rec1.shape[0]):
        af2_wt_rmsd_list_rec1[idx,j] = ls[idx]

af2_wt_rmsd_list_rec1 = np.asarray(af2_wt_rmsd_list_rec1)
```

```python
af2_wt_rmsd_list_rec1
np.mean(af2_wt_rmsd_list_rec1, axis=1)
```

```python
# Create lists for the plot
uniprot_ids = ['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5']
x_pos = np.arange(len(uniprot_ids))
means = np.mean(af2_wt_rmsd_list_rec1, axis=1).tolist()
errors = np.std(af2_wt_rmsd_list_rec1, axis=1).tolist()
errors
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, means, yerr=errors, align='center', alpha=1, ecolor='black', capsize=10)
ax.set_ylabel('RMSD [Å]')
ax.set_xticks(x_pos)
ax.set_xticklabels(uniprot_ids)
ax.set_title("P61626: Mean RMSD between ESMFold Model and AF2's 5 Models")
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('/Users/holger/Desktop/master_thesis/notebooks/figs/bar_plot_P61626_ESM_AF2_rec1.png')
plt.show()
```
